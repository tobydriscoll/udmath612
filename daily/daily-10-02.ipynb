```julia
using LinearAlgebra
```

```julia
function mgs(A)
    m,n = size(A)
    Q = zeros(eltype(A),m,n)
    R = zeros(eltype(A),n,n)
    B = copy(A)
    for k = 1:n
        R[k,k] = norm(B[:,k])
        Q[:,k] = B[:,k]/R[k,k]
        for j = k+1:n
            R[k,j] = dot(Q[:,k],B[:,j])
        end
        B -= Q[:,k]*R[k,:]'
    end
    return Q,R
end  
```
