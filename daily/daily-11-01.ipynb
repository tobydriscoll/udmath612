Looking into Richardson iteration as in TB problem 35.3.

```julia
using LinearAlgebra, PyPlot

m = 500
A = 2I + 0.5*randn(m,m)/sqrt(m);
theta = pi*(0:m-1)/(m-1)
d = @. (-2+2*sin(theta)) + 1im*cos(theta)
D = Diagonal(d)
A = A + D
lam = eigvals(A)
plot(real(lam),imag(lam),".")
axis("equal") 
title("Eigenvalues of A");
```

```julia
b = normalize(randn(m))
x = zero(b)
# nr = zeros(1000)
# α = 4/3  # "optimal" choice
# for n = 1:1000
#     x = x + α*(b-A*x)
#     nr[n] = norm(b-A*x)
# end
# semilogy(1:1000,nr,"-o")

α = .01   # "tiny" choice
nr = zeros(10000)
x = zero(b)
for n = 1:10000
    x = x + α*(b-A*x)
    nr[n] = norm(b-A*x)
end
semilogy(1:10000,nr,"-o")
```

```julia
b = normalize(randn(m))
x = zero(b)
nr = zeros(1000)
α = 4/3  # "optimal" choice
for n = 1:1000
    x = x + α*(b-A*x)
    nr[n] = norm(b-A*x)
end
semilogy(1:1000,nr,"-o")
```
