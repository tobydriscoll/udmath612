```julia
using SpecialFunctions, LinearAlgebra, PyPlot
```

```julia
pwd()
```

```julia
X=imread("Hello.png");
```

```julia
A=float(X[:,:,1]);
```

```julia
size(A)
```

```julia
maximum(A)
```

```julia
U,s,V = svd(A);
semilogy(s,"o")
```

```julia
Ak(k) = U[:,1:k]*Diagonal(s[1:k])*V[:,1:k]'
```

```julia
imshow(Ak(8),cmap="gray")
```

```julia
norm(A-Ak(8))/norm(A)
```

```julia
(norm(A)-norm(Ak(8)))/norm(A)
```
