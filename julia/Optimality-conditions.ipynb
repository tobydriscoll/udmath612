# Optimality conditions

## Necessary conditions and curvature

For a smooth $f$ to have a local minimum at $r$, it must be that $\nabla f(r)=0$ and the Hessian is positive semidefinite at $r$. The Hessian essentially is a curvature term. If the curvature is negative (North Pole ice cap) or a saddle, then the critical point is not a local min.

```julia
using Plots # alternative graphics package
gr(leg=false)
```

```julia
H = [2 -1;-1 -1]  # indefinite Hessian -> saddle
f(x) = 1 + x'*H*x
x = -1:0.02:1
contour(x,x,(x,y)->f([x,y]),leg=true)
scatter!([0],[0],c=:red,m=:star)
```

```julia
plotlyjs()  # interactive graphics
surface(x,x,(x,y)->f([x,y]))
gui();
```

```julia
gr()
H = [2 -1;-1 3]  # PD Hessian -> local min
contour(x,x,(x,y)->f([x,y]))
plot!([0],[0],m=:star,c=:red,leg=true)
```

## Sufficient conditions 

However, to *guarantee* a local minimum (i.e. sufficient conditions), the Hessian must be positive semidefinite. Without this, a zero direction for the Hessian can be bent downward by higher-order terms.

```julia
H = [2 -1;-1 0]  # semidefinite Hessian 
f(x) = 1 + x'*H*x - 10*x[2]^4
x = -0.5:0.01:0.5
plotlyjs()  # interactive graphics
surface(x,x,(x,y)->f([x,y]),xaxis="x",yaxis="y")
gui();
```

```julia
f([0,0])
```

```julia
f([0,.01])
```
